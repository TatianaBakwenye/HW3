```python
# c. Remove those rows that contain NaN values in the columns: age, gender, ethnicity.
def drop_na(df, columns):
    df_drop = df.dropna(subset=columns)
    return df_drop
```

```python
result_c = drop_na(df, ['age', 'gender', 'ethnicity'])
print(result_c.shape)
print(result_c.describe())
```

```python
# d. Fill NaN with the mean value of the column in the columns: height, weight.
def fill_na(df, columns):
    df[columns] = df[columns].apply(lambda col: col.fillna(col.mean()))
    return df
```

```python
result_d = fill_na(df, ['height', 'weight'])
print(result_d.shape)
print(result_d.describe())
```

```python
# Clean data from NA (ex. c and d)
def clean_na(df, columns_drop, columns_fill):
    df_clean = df.dropna(subset=columns_drop)
    df_clean[columns_fill] = df_clean[columns_fill].apply(lambda col: col.fillna(col.mean()))
    
    return df_clean
```

```python
columns_drop = ['age', 'gender', 'ethnicity']
columns_fill = ['height', 'weight']
result_clean = clean_na(df, columns_drop, columns_fill)
result_clean.describe()
#result_clean.shape
```

```python
# e. Generate dummies for ethnicity column (One hot encoding).
def one_hot(df, columns):
    encoder = OneHotEncoder()
    encoded_data = encoder.fit_transform(df[columns]).toarray()
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns))
    df.reset_index(drop=True, inplace=True)  # Reset the index and drop the old index column
    encoded_df = pd.concat([df, encoded_df], axis=1)
    return encoded_df
```

```python
result_e = one_hot(df, ['ethnicity'])
print(result_e.head())
```

```python
def binary(df, columns):
    df['Binary'] = df[columns].apply(lambda x: 1 if x == 'M' else 0)
    return df
```

```python
result_f = binary(df, 'gender')
print(result_f.head())
```

```python
# Encode variables (ex. e and f)
def encode(df, columns_hot, columns_binary):
    encoder = OneHotEncoder()
    encoded_data = encoder.fit_transform(df[columns_hot]).toarray()
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_hot))
    df.reset_index(drop=True, inplace=True)  # Reset the index and drop the old index column
    encoded_df = pd.concat([df, encoded_df], axis=1)
    df['Binary'] = df[columns_binary].apply(lambda x: 1 if x == 'M' else 0)
    return encoded_df
    
```

```python
result_encoded = encode(df, ['ethnicity'], 'gender')
print(result_encoded.head())
```

```python
def process_data(df, columns_drop, columns_fill, columns_hot, columns_binary):
    df_clean = df.dropna(subset=columns_drop)
    df_clean[columns_fill] = df_clean[columns_fill].apply(lambda col: col.fillna(col.mean()))
    df_clean['Binary'] = df_clean[columns_binary].apply(lambda x: 1 if x == 'M' else 0)
    encoder = OneHotEncoder()
    encoded_data = encoder.fit_transform(df_clean[columns_hot]).toarray()
    encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_hot))
    df_clean.reset_index(drop=True, inplace=True)  # Reset the index and drop the old index column
    encoded_df = pd.concat([df_clean, encoded_df], axis=1)
    
    return encoded_df
```

```python
columns_drop = ['age', 'gender', 'ethnicity']
columns_fill = ['height', 'weight']
columns_hot = ['ethnicity']
columns_binary = 'gender'
result_processed = process_data(df, columns_drop, columns_fill, columns_hot, columns_binary)
result_processed.head()
```
